```python
import xarray as xar
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
ds = xar.open_dataset('ds_nu1e1.nc')
```

```python
vort = ds.vorticity


vort_t = vort[:,256,256]
t = vort.t
```

```python
plt.plot(t, vort_t)
plt.show()
```

```python
# Read all data sets using glob to find pathnames (returns array of strings with path)
files = sorted(glob.glob('*.nc'))

# names = ['ds_nu0p1e1', 'ds_nu1e1', 'ds_nu1e2', 'ds_nu1e3', 'ds_nu1e4', 'ds_nu1e5']

files
```

```python
ds_nu0p1e1 = xar.open_dataset(files[0])
ds_nu1e1 = xar.open_dataset(files[1])
ds_nu1e2 = xar.open_dataset(files[2])
ds_nu1e3 = xar.open_dataset(files[3])
ds_nu1e4 = xar.open_dataset(files[4])
ds_nu1e5 = xar.open_dataset(files[5])

ds_nu0p1e1_vort = ds_nu0p1e1.vorticity[:,256,256]
ds_nu1e1_vort = ds_nu1e1.vorticity[:,256,256]
ds_nu1e2_vort = ds_nu1e2.vorticity[:,256,256]
ds_nu1e3_vort = ds_nu1e3.vorticity[:,256,256]
ds_nu1e4_vort = ds_nu1e4.vorticity[:,256,256]
ds_nu1e5_vort = ds_nu1e5.vorticity[:,256,256]

t = ds_nu1e5_vort.t
```

```python
plt.figure()
plt.plot(t, ds_nu0p1e1_vort, label='nu1e-1')
plt.plot(t, ds_nu1e1_vort, label='nu1e1')
plt.plot(t, ds_nu1e2_vort, label='nu1e2')
plt.plot(t, ds_nu1e3_vort, label='nu1e3')
plt.plot(t, ds_nu1e4_vort, label='nu1e4')
plt.plot(t, ds_nu1e5_vort, label='nu1e5')

plt.legend()
plt.title('Vorticity for varying magntiudes of diffusion')
plt.xlabel('time (t)')
plt.ylabel('vorticity')

plt.show()
```
